```python
import pandas as pd
import numpy as np
from transformers import AutoTokenizer, AutoModelForSequenceClassification, Trainer, TrainingArguments
from datasets import Dataset
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, precision_recall_fscore_support
import torch
import os

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Device used: {device}")

train_file = "training_data.csv"
val_file = "validation_data.csv"

if not os.path.exists(train_file) or not os.path.exists(val_file):
    print(f"Error: One of the data files ({train_file} or {val_file}) is missing.")
    exit()

try:
    train_df = pd.read_csv(train_file)
    val_df = pd.read_csv(val_file)
except Exception as e:
    print(f"Error loading data: {e}")
    exit()

required_columns = ["text", "label"]
if not all(col in train_df.columns for col in required_columns) or not all(col in val_df.columns for col in required_columns):
    print("Error: 'text' or 'label' column missing in one of the files.")
    exit()

label_encoder = LabelEncoder()
try:
    train_df["label"] = label_encoder.fit_transform(train_df["label"])
    val_df["label"] = label_encoder.transform(val_df["label"])
    print("Encoded labels:", list(label_encoder.classes_))
    print("Corresponding IDs:", list(range(len(label_encoder.classes_))))
except Exception as e:
    print(f"Error encoding labels: {e}")
    exit()

np.save("label_encoder.npy", label_encoder.classes_)

train_dataset = Dataset.from_pandas(train_df)
val_dataset = Dataset.from_pandas(val_df)

model_name = "aubmindlab/bert-base-arabertv2"
try:
    tokenizer = AutoTokenizer.from_pretrained(model_name)
    model = AutoModelForSequenceClassification.from_pretrained(model_name, num_labels=len(label_encoder.classes_))
    model.to(device)
except Exception as e:
    print(f"Error loading model or tokenizer: {e}")
    exit()

def tokenize_function(examples):
    return tokenizer(examples["text"], padding="max_length", truncation=True, max_length=128)

try:
    train_dataset = train_dataset.map(tokenize_function, batched=True)
    val_dataset = val_dataset.map(tokenize_function, batched=True)
except Exception as e:
    print(f"Error during tokenization: {e}")
    exit()

train_dataset = train_dataset.rename_column("label", "labels")
val_dataset = val_dataset.rename_column("label", "labels")
train_dataset = train_dataset.with_format("torch", columns=["input_ids", "attention_mask", "labels"], output_all_columns=True)
val_dataset = val_dataset.with_format("torch", columns=["input_ids", "attention_mask", "labels"], output_all_columns=True)

def compute_metrics(eval_pred):
    logits, labels = eval_pred
    predictions = np.argmax(logits, axis=-1)
    accuracy = accuracy_score(labels, predictions)
    precision, recall, f1, _ = precision_recall_fscore_support(labels, predictions, average="weighted")
    return {
        "accuracy": accuracy,
        "precision": precision,
        "recall": recall,
        "f1": f1
    }

training_args = TrainingArguments(
    output_dir="./results",
    eval_strategy="epoch",
    learning_rate=2e-5,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    num_train_epochs=15,
    weight_decay=0.01,
    logging_dir="./logs",
    logging_steps=10,
    save_strategy="epoch",
    load_best_model_at_end=True,
    metric_for_best_model="f1",
    warmup_steps=500,
    report_to="none"
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset,
    compute_metrics=compute_metrics
)

print("Starting training...")
try:
    trainer.train()
except Exception as e:
    print(f"Error during training: {str(e)}")
    print("Error details:", e.__class__.__name__)
    exit()

print("Saving model...")
try:
    model.save_pretrained("./trained_model")
    tokenizer.save_pretrained("./trained_model")
    print("Model and tokenizer saved successfully!")
except Exception as e:
    print(f"Error saving model: {e}")
    exit()

print("Evaluating final model...")

def safe_evaluate(trainer, dataset):
    dataloader = trainer.get_eval_dataloader(dataset)
    output = trainer.prediction_loop(dataloader, description="Evaluation")
    logits = output.predictions
    labels = output.label_ids
    return compute_metrics((logits, labels))

try:
    eval_results = safe_evaluate(trainer, val_dataset)
    print("Final evaluation results:", eval_results)
except Exception as e:
    print(f"Evaluation failed: {str(e)}")
    exit()
```

```python
import sqlite3
from datetime import datetime
import numpy as np
import torch
from transformers import AutoModelForSequenceClassification, AutoTokenizer, pipeline


def init_db():
    conn = sqlite3.connect("chat_history.db")
    c = conn.cursor()
    c.execute('''
        CREATE TABLE IF NOT EXISTS chat_history (
            id INTEGER PRIMARY KEY AUTOINCREMENT,
            user_id TEXT NOT NULL,
            text TEXT NOT NULL,
            label TEXT NOT NULL,
            timestamp TEXT NOT NULL
        )
    ''')
    conn.commit()
    conn.close()


def save_interaction(user_id, text, label):
    conn = sqlite3.connect("chat_history.db")
    c = conn.cursor()
    timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    c.execute("INSERT INTO chat_history (user_id, text, label, timestamp) VALUES (?, ?, ?, ?)",
              (user_id, text, label, timestamp))
    conn.commit()
    conn.close()


def get_user_history(user_id, limit=10000):
    conn = sqlite3.connect("chat_history.db")
    c = conn.cursor()
    c.execute("SELECT text, label, timestamp FROM chat_history WHERE user_id = ? ORDER BY timestamp DESC LIMIT ?",
              (user_id, limit))
    history = c.fetchall()
    conn.close()
    return [{"text": row[0], "label": row[1], "timestamp": row[2]} for row in history]


model_path = "./trained_model"
try:
    tokenizer = AutoTokenizer.from_pretrained(model_path)
    model = AutoModelForSequenceClassification.from_pretrained(model_path)
except Exception as e:
    print(f"Error loading model: {e}")
    exit()


device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)


try:
    label_encoder = np.load("label_encoder.npy", allow_pickle=True)
except FileNotFoundError:
    print("Error: label_encoder.npy not found. Ensure training completed successfully.")
    exit()

classifier = pipeline("text-classification", model=model, tokenizer=tokenizer, top_k=None, device=0 if torch.cuda.is_available() else -1)


def classify_intent(user_input):
    results = classifier(user_input)[0]
    max_score = max(results, key=lambda x: x['score'])
    label = max_score['label']
    try:
        label_id = int(label.replace("LABEL_", ""))
        intent = label_encoder[label_id]
    except ValueError:
        intent = "unknown"
    return intent


def chatbot_response(user_input, user_id):
    user_input = user_input.strip().lower()
    intent = classify_intent(user_input)


    save_interaction(user_id, user_input, intent)


    history = get_user_history(user_id)
    contextual_message = ""


    if intent == "greeting":
        base_response = ("وعليكم السلام! أنا خضر، هنا لمساعدتك في زراعة الأشجار. "
                         "ما الذي تريد مناقشته؟ (زراعة، أمراض، أنواع للشوارع، العناية والسماد)")
        if history:
            last_interaction = history[0]
            contextual_message = (f"آخر مرة تكلمنا عن {last_interaction['label']}، "
                                  f"دلوقتي عايز/ة تبدأ من جديد؟ {base_response}")
        else:
            contextual_message = base_response
    elif intent == "planting":
        base_response = ("لزراعة شجرة: \n"
                         "1. اختر مكانًا به شمس كافية وتربة جيدة التصريف.\n"
                         "2. احفر حفرة بعمق وعرض ضعف حجم جذور الشجرة.\n"
                         "3. ضع الشجرة في الحفرة، ثم أضف التربة وادكها برفق.\n"
                         "4. اسقِ الشجرة جيدًا وأضف طبقة من المهاد (مثل القش) لحفظ الرطوبة.")
        if history and history[0]["label"] == "planting":
            contextual_message = (f"يبدو إنك مهتم/ة بـزراعة الأشجار! تحدثنا عنها قبل كده. "
                                  f"إليك نصيحة جديدة: {base_response}")
        else:
            contextual_message = base_response
    elif intent == "diseases":
        if "البياض الدقيقي" in user_input:
            base_response = "البياض الدقيقي: بقع بيضاء على الأوراق، العلاج: مبيدات فطرية."
        elif "العفن الجذري" in user_input:
            base_response = "العفن الجذري: بسبب الري الزائد، الحل: تحسين التصريف."
        elif "تسوس الجذع" in user_input:
            base_response = "تسوس الجذع: بسبب الفطريات، الحل: تقليم الأجزاء المصابة."
        else:
            base_response = ("من أشهر أمراض الأشجار:\n"
                             "- العفن الجذري: بسبب الري الزائد، الحل: تحسين التصريف.\n"
                             "- البياض الدقيقي: بقع بيضاء على الأوراق، العلاج: مبيدات فطرية.\n"
                             "- تسوس الجذع: بسبب الفطريات، الحل: تقليم الأجزاء المصابة.\n"
                             "هل تريد تفاصيل عن مرض معين؟")
        if history and history[0]["label"] == "diseases":
            contextual_message = (f"يبدو إنك مهتم/ة بأمراض الأشجار! تحدثنا عنها قبل كده. "
                                  f"إليك معلومات إضافية: {base_response}")
        else:
            contextual_message = base_response
    elif intent == "tree_types":
        base_response = ("أنواع مناسبة للشوارع:\n"
                         "- الزنزلخت: سريع النمو، ظل كثيف، يتحمل الجفاف.\n"
                         "- الكافور: أوراق عطرية، مناسب للمدن.\n"
                         "- النخيل: جمالي ويتحمل الحرارة.\n"
                         "- الأكاسيا: قوية ومناسبة للمناطق الجافة.\n"
                         "اختر نوعًا للشارع بناءً على مناخ منطقتك!")
        if history and history[0]["label"] == "tree_types":
            contextual_message = (f"يبدو إنك مهتم/ة بأنواع الأشجار! تحدثنا عنها قبل كده. "
                                  f"إليك نصيحة جديدة: {base_response}")
        else:
            contextual_message = base_response
    elif intent == "care":
        if "كيف أعتني" in user_input:
            base_response = "العناية بالأشجار تتطلب اهتمامًا دقيقًا ومستمرًا لضمان نموها بشكل صحي وقوي..."
        elif "متى أقلم" in user_input or "تقليم" in user_input:
            base_response = "التقليم عملية أساسية للحفاظ على صحة الشجرة وجمالها..."
        elif "ري" in user_input or "كم مرة أسقي" in user_input:
            base_response = "الري المنتظم يعتمد على عمر الشجرة وحالة المناخ..."
        elif "تربة" in user_input:
            base_response = "الاهتمام بالتربة يشمل فحصها دوريًا..."
        elif "سماد" in user_input or "ما السماد" in user_input:
            base_response = "السماد المناسب يختلف حسب احتياجات الشجرة..."
        else:
            base_response = "العناية بالأشجار تشمل الري المنتظم، التقليم، والتسميد..."
        if history and history[0]["label"] == "care":
            contextual_message = (f"يبدو إنك مهتم/ة بالعناية بالأشجار! تحدثنا عنها قبل كده. "
                                  f"إليك نصيحة جديدة: {base_response}")
        else:
            contextual_message = base_response
    elif intent == "farewell":
        base_response = "مع السلامة! أتمنى أن أكون قد ساعدتك في زراعة أشجارك."
        contextual_message = base_response
    else:
        base_response = ("آسف، لم أفهم. يمكنك السؤال عن (طرق الزراعة، أمراض الأشجار، أنواع للشوارع، العناية والسماد). "
                         "ما الذي تريد مناقشته؟")
        contextual_message = base_response

    return contextual_message


init_db()


print("مرحبًا! أنا خضر، شات بوت عربي متخصص في زراعة الأشجار. اسألني عن الزراعة، الأمراض، الأنواع المناسبة للشوارع، أو العناية والسماد!")
user_id = input("من فضلك، أدخل معرف المستخدم الخاص بك (مثل user1): ")
while True:
    user_input = input("أنت: ")
    response = chatbot_response(user_input, user_id)
    print(f"خضر: {response}")
    if user_input.lower() in ["وداعًا", "مع السلامة", "خروج"]:
        break
```
